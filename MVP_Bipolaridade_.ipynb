1.Definição do problema:
	
O conjunto de dados "Bipolar vs Non-Bipolar Handwriting" é um problema de classificação binária, onde precisamos classificar se uma amostra de escrita manual pertence a um indivíduo bipolar ou não bipolar.

1.Objetivo:
	
O objetivo é treinar um modelo de classificação capaz de identificar caracteriscticas distintivas em uma amostra de escrita manual para indicar se pertence a um indivíduo com transtorno bipolar ou não. Isso pode ajudar a fornecer insights adicionais sobre as características comportamentais e neurológicas do transtorno, que deve apoiar o diagnóstico médico em sua eficiência e a eficácia, elevando a qualidade dos cuidados de saúde dos pacientes.

O conjunto de dados contém um total de 1250 amostras, sendo 500 amostras de indivíduos com transtorno bipolar e 750 amostras de indivíduos sem transtorno bipolar. Cada amostra é representada como uma imagem de escrita manual de tamanho 300x300 pixels.

```python
# Vamos verificar dados faltantes:

import pandas as pd

# carregar o dataset em um dataframe
df = pd.read_csv('Original_Dataset.csv')

# identificar valores nulos em cada coluna
print(df.isnull().sum())
```

A saida apresentada mostra que não há valores faltantes em nosso conjunto de dados.

```python
# Verificando se temos valores duplicados em nossos dados

df.duplicated()
```

A nossa saida acima mostra que nao temos valores duplicados

```python
import seaborn as sns
import matplotlib.pyplot as plt

# criando um boxplot para cada coluna numérica do dataframe para que possamos identificar a presença de outliers.
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(4, 2))
    sns.boxplot(x=df[col])
    plt.show()
```

Analisando os nossos graficos podemos perceber que temos dados regulares

```python
# Utilizando o método describe para obter um resumo estatistico dos dados numéricos

df.describe()
```

Verificamos acima que não há valores ausentes em nenhum dos atributos numéricos e nao foi identificada nenhuma inconsistência nos valores.
A questão de a média e a mediana terem valores próximos indica que as distribuições dos atributos são aproximadamente simétricas.

```python
# Verificando o formato dos valores numericos
num_cols = df.columns[:-2]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Substituindo os valores faltantes pela mediana

df = df.fillna(df.median())
```

A padronização é utilizada para transformar os dados em uma escala com média igual a zero e desvio-padrão igual a um.

```python
# Padronizando

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Exibindo
df
```
